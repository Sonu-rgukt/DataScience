```python
import matplotlib.pyplot as plt
import numpy as np
plt.title("line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
xpoints = np.array([10,20])
ypoints = np.array([40,70])
plt.plot(xpoints,ypoints)
plt.show
```
